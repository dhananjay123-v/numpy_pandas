# a) Create a Series using pandas and display

```python
import pandas as pd
# Creating a Pandas Series
data = [10, 20, 30, 40] 
series = pd.Series(data)
print("Pandas Series:")
print(series)
```

# b) Access the index and the values of the Series

```python
print("Index:", series.index)
print("Values:", series.values)
```

# c) Compare a NumPy array with a Pandas Series

```python
import numpy as np
array = np.array([10, 20, 30, 40]) 
series = pd.Series([10, 20, 30, 40])
print("Numpy Array:\n", array) 
print("Pandas Séries:\n", series)
```

# d) Define Series objects with individual (custom) indices

```python
marks = pd.Series([85, 90, 75, 88],
index=["Math", "Physics", "Chemistry", "English"])
print("Marks Series with Subject Index:\n", marks)
```

# e) Access a single value of a Series

```python
print("Marks in Physics:", marks["Physics"])
```

# g) Usage of different methods in matplotlib

```python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
# Line Plot
plt.plot(x, y, label='Line Plot', color='blue', marker='o')
plt.title("Sample Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.grid(True)
plt.show()
# Bar Chart
plt.bar(x, y, color='green')
plt.title("Sample Bar Chart")
plt.show()
# Pie Chart
labels = ['A', 'B', 'C', 'D']
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title("Sample Pie Chart")
plt.show()
```
