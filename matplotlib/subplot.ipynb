```python
import matplotlib.pyplot as plt

z=[1,2,3,4,5]
m=[10,20,25,15,40]

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(z,m)
plt.title('Line chart')
plt.xlabel('Sale per Unit')

plt.subplot(2,2,2)
plt.pie(m,labels=z,autopct='%1.1f%%')
plt.title('pie chart')

plt.subplot(2,2,4)
plt.bar(z,m,color='green')
plt.title('Bar chart')
plt.xlabel('Sale per Unit')

plt.subplot(2,2,3)
core=[20,36,52,21,26,28,56,36,45,55,52,44,41,49,36,52,21]
plt.hist(core,bins=5,color='orange',edgecolor='black')
plt.title('Hiat chart')
plt.xlabel('Sale per Unit')


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.figure(figsize=(10, 8))  # Set figure size

# Plot 1 - Line chart
plt.subplot(2, 2, 1)  # (rows, cols, index)
plt.plot(x, y, color='blue')
plt.title('Line Plot')

# Plot 2 - Bar chart
plt.subplot(2, 2, 2)
plt.bar(x, y, color='green')
plt.title('Bar Chart')

# Plot 3 - Scatter plot
plt.subplot(2, 2, 3)
plt.scatter(x, y, color='red')
plt.title('Scatter Plot')

# Plot 4 - Pie chart
plt.subplot(2, 2, 4)
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title('Pie Chart')

plt.tight_layout()  # Prevent overlap
plt.show()
```
