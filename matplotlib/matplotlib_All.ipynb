```python
# Line Chart
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data_3000.csv')
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')

plt.figure(figsize=(12, 6))
plt.plot(df['date'], df['sales'], marker='o', linestyle='-', color='blue')
plt.title('Sales Over Time - Line Chart')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#bar chart
plt.figure(figsize=(12, 6))
plt.bar(df['date'], df['sales'], color='orange')
plt.title('Daily Sales - Bar Chart')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#area chart

plt.figure(figsize=(12, 6))
plt.fill_between(df['date'], df['sales'], color='skyblue', alpha=0.5)
plt.plot(df['date'], df['sales'], color='blue')
plt.title('Sales Over Time - Area Chart')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#acatter plot 

plt.figure(figsize=(10, 6))
plt.scatter(df['date'], df['sales'], 
            color='green',
            marker='^'
           )
plt.title('Sales Scatter Plot')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#boz plot

plt.figure(figsize=(6, 6))
plt.boxplot(df['sales'])
plt.title('Sales Distribution')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# year 2023

import pandas as pd

df = pd.read_csv('sales_data_3000.csv')
df['date'] = pd.to_datetime(df['date'])

# Filter for a specific year, e.g., 2023
df_2023 = df[df['date'].dt.year == 2023]

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_2023['date'], df_2023['sales'], color='blue', marker='o')
plt.title('Sales Over Time - 2023')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df_2023['date'], df_2023['sales'], color='orange')
plt.title('Sales - Bar Chart for 2023')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and parse dates
df = pd.read_csv('sales_data_3000.csv')
df['date'] = pd.to_datetime(df['date'])

# Extract year
df['year'] = df['date'].dt.year

# Group by year and sum sales
sales_by_year = df.groupby('year')['sales'].sum()

# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(sales_by_year, labels=sales_by_year.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Sales Distribution by Year')
plt.axis('equal')  # Make it a circle
plt.show()
```
